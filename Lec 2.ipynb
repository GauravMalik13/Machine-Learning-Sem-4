List is Mutable while Tuples are immutable. In simple words List can be appended or modified whereas tuples can not be modified.

````python
l = [1, 2, 3]
t = (2, 4, 6)
l.append(3)
print(f"list is Mutable, New list: {l}")
print(f"Tuple is immutable, {t}")
````

# Lists inside tuples are not immutable.

````python
t1 : tuple[int, list[int | str]] = (2, 3, 5, ["hello", "what", "is", "your"])
print(f"Before Modifying : {t1}")

t1[-1][-1] = "name"

print(f"After Modifying: {t1}")
````

````python
import numpy as np
import matplotlib.pyplot as plt
def coulomb_force(r: float, *, q1: float, q2: float, k: float= 9e9):
    return (k*q1*q2)/r**2

r = np.linspace(0.01, 0.1, 100)
f = coulomb_force(r, q1 = 1e-3, q2 = 1e-8)
plt.plot(r, f)
plt.title("Coulomb Force Vs Distance")
plt.show()
````

### Dict

````python
particle : dict[str, str | float]= {
    "name" : "proton",
    "mass" : 1.67e-27,
    "charge": 1.6e-19,
}
print(particle)
print(f"Dict Items : {particle.items()}")
print(f"Dict Keys : {particle.keys()}")
print(f"Dict Values: {particle.values()}")
````

### List and Dict Comprehensions

````python
first_Name = ["Gaurav", "Pushkaraj", "Gagan"]
last_Name = ["Malik", "Rode", "Bishnoi"]

name = [first + " " + last_Name[i] for i, first in enumerate(first_Name)]
print(name)

name_dict : dict[str, str] = { first : last_Name[i] for i , first in enumerate(first_Name) if i < 2}
print(name_dict)
````

### Lambda Functions

````python
filtered_list = list(filter(lambda x : len(x)>12, name))
print(filtered_list)
````

## Libraries/Module

Different Libraries can be used to do different stuff.

To import a module/library, we use the `import` statement

> ```python
> import module
>
> (or)
>
> import module as alias
> ```

Here, `module` is the name of the module and `alias` is the keyword that the user defines to call the objects in that module.

Note: `Alias` is not necessary but is recomended.
